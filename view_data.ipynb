```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

```python
print(df[df.isnull().any(axis=1)])
```

```python
df.describe()
```

```python
id_list = df["world_idx"].unique()
```

```python
# find world idx 0-299 not in id_list
missing = [i for i in range(300) if i not in id_list]
print(missing)
```

```python
# add timestep column is the index of the row but reset to 0 every new world idx
df['timestep'] = df.groupby('world_idx').cumcount()
df
```

```python
# sort by world_idx and timestep
df_sorted = df.sort_values(by=['world_idx', 'timestep'])
df_sorted
```

```python
# save to csv
df_sorted.to_csv('data_sorted.csv', index=False)
```
